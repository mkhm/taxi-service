```python
%matplotlib inline
import pandas as pd
import matplotlib
from matplotlib import animation, colors
import numpy as np
import pylab as plt
from time import time, sleep

def distributionHistogram(data, numberOfBin, center=True):
    if type(data) != np.ndarray:
        data = np.array(data)
    y, x = np.histogram(data, numberOfBin, density=True)
    N = len(data)
    if center:
        x = (x[:-1] + x[1:])/2
    return (x, y)
```

```python
def plotCorrelationMatix(data, normalize=True, column_list=column_list, saveName=-1, **keyargs):
    fig = plt.figure(figsize=(700/80,700/80))
    ax = fig.add_subplot(1, 1, 1)
    ax.xaxis.set_ticks_position('top')
    ##ax.yaxis.grid(linestyle = '-', color = 'gray')
    ##ax.set_title('asdasdasdasdasdasd')
    
    aa=ax.imshow(data, cmap=plt.cm.hot) ##, cmap=plt.cm.hot {hot, autumn, Blues, spring, summer, winter, gray, cool}
                                         ##, interpolation='nearest' {'none', 'nearest', 'bilinear', 'bicubic'}
    ax.set_yticks(np.arange(len(column_list)))
    ax.set_xticks(np.arange(len(column_list)))
    ax.set_yticklabels(column_list, **keyargs)
    ax.set_xticklabels(column_list, rotation='vertical', **keyargs)
    
    if normalize == True:
        norm = colors.Normalize(vmin=-1.0, vmax=1.0)
        aa.set_norm(norm)
    
    plt.colorbar(aa)
    plt.show()
    
    if saveName != -1:
        fig.savefig(saveName, bbox_inches='tight', pad_inches=0.1, dpi=150) # , dpi=80, transparent=True

```

```python
def plotDistributionOfField(columnName, NumberOfRemoveFromEnd=0, bins=50, figsize=(12,12), basex=1, basey=1, **kwargs):
    ff = excel[columnName].copy()
    fff = np.asanyarray(ff)
    fff.sort()
    if NumberOfRemoveFromEnd != 0:
        fff = fff[:-NumberOfRemoveFromEnd]
#     print fff
    # print max(fff)
    # a=plt.hist(fff, 150);

    a=np.histogram(fff, bins, density=True);
    print a[0], a[1]
    plt.figure(figsize=figsize)
    plt.plot(a[1][:-1], a[0], **kwargs)
#     print plt.xlim()
    if basex!=1:
        plt.xscale('log', basex=basex) # basex=4
    if basey!=1:
        plt.yscale('log', basey=basey) # basey=4
```
